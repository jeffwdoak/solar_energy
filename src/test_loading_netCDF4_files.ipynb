## Create DataFrame out of geography data
I will need to look at the other data types and see how to combine them all together.

Latitude and Longitude could be used as a MultiIndex, at least for this dataset.

```python
def netcdf4_to_dataframe(netcdf):
    data_as_array = np.array([np.array(i).flatten() for i in netcdf.variables.values()]).T
    data_as_dataframe = pd.DataFrame(data_as_array, columns=netcdf.variables.keys())
    return data_as_dataframe
    
```
